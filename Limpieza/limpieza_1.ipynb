```python
df[['cons,price,idx', 'cons,conf,idx', 'euribor3m']] = df[['cons,price,idx', 'cons,conf,idx', 'euribor3m']].replace(',', '.', regex = True).astype('float64', copy=False, errors = 'raise')
```

```python
df["poutcome"] = df["poutcome"].apply(lambda elemento : elemento.lower())
```

```python
def minusculas(elemento):

    if isinstance(elemento, str):

        return elemento.lower()
    else:
        return elemento
```

```python
df["marital"] = df["marital"].apply(minusculas)
```
